```python
import pandas as pd
import numpy as np

# Raw data dictionary
data = {
    'Order_ID': [1001, 1002, 1003, 1004, 1005, 1005, 1006, 1007, 1008, 1009, 1010, 1011],
    ' Region ': ['north', 'South', 'EAST', 'west', 'North', 'North', 'south', 'East', 'WEST', 'north', 'South', 'east'],
    'Category': ['Electronics', 'electronics', 'Furniture', 'furnture', 'Electronics', 'Electronics', 'Office Supplies', 'office supplies', 'Electronics', 'Furniture', 'Electronics', 'Electronics'],
    'Sales': ['$1200', '$850', '$2500', '$1750', '$1200', '$1200', '$300', '$450', '$1100', np.nan, '$950', '$1300'],
    'Quantity': [5, 3, 2, 1, 5, 5, 7, 4, 6, 2, np.nan, 4],
    'Profit': [200, 120, 400, 250, 200, 200, 50, 60, 180, 90, 150, 220],
    'Order_Date': ['01-01-2024', '02-01-2024', '03-01-2024', '04-01-2024', '01-01-2024', '01-01-2024', '05-01-2024', '06-01-2024', '07-01-2024', '08-01-2024', '09-01-2024', '10-01-2024']
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
df = df.drop_duplicates()
```

```python
df['region'] = df['region'].str.title()
df['category'] = df['category'].str.title()
```

```python
df['category'] = df['category'].replace('Furnture','Furniture')
```

```python
df['sales'] = df['sales'].str.replace('$','',regex = False)
```

```python
df['sales'] = pd.to_numeric(df['sales'])
```

```python
df['sales'].fillna(df['sales'].mean(),inplace = True)
```

```python
df['quantity'].fillna(df['quantity'].mean(), inplace=True)
```

```python
df['sales'] = df['sales'].astype(float)
df['quantity'] = df['quantity'].astype(int)
df['profit'] = df['profit'].astype(float)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.info()
```

```python
high_sales = df[df['sales'] > 1000]
high_sales
```

```python
region_sales = df.groupby('region')['sales'].sum()
print(region_sales)
```

```python
category_profit = df.groupby('category')['profit'].sum()
print(category_profit)
```
